```python
import datetime
import pandas as pd
    
```

```python
# chargement du fichier
data = pd.read_csv("MLAIM.csv")
```

```python
# les valeurs nulles
null_values = data.isnull().sum()
```

```python
#   Suppression des lignes nulles
clean_data = data.dropna(axis=0, how="any").reset_index(drop=True)
```

```python
# Suppression des valeurs dupliquées
clean_data = clean_data.drop_duplicates().reset_index(drop=True)
```

```python
# Unification de la date de naissance
clean_data["Date_Naissance"] = pd.to_datetime(clean_data["Date_Naissance"], format="mixed")
```

```python
# calcule d'age
date_separated = clean_data["Date_Naissance"].astype(str).str.split("-")
for i in range(len(date_separated)):
    today = datetime.datetime.now()
    calculated_age = datetime.datetime.now().year - int(date_separated.iloc[i][0])
    if (today.month < int(date_separated.iloc[i][1])) and (today.day < int(date_separated.iloc[i][2])):
        calculated_age -= 1
    clean_data.loc[i, "age"] = calculated_age
```

```python
#   Strings concatenation Function
def concatenate_strings(str):
    if "-" in str:
        str_list = list(str)
        str_list.pop(str.find("-"))
        str = "".join(str_list)
    str = str.split(' ')
    for i in range(len(str)):
        str = "".join(str)
    return str
```

```python
#   Création de la colonne Email
email_template =  "@usmba.ac.ma"
for i in range(len(clean_data)):
    prenom = concatenate_strings(clean_data["Prenom"].iloc[i])
    nom = concatenate_strings(clean_data["Nom"].iloc[i])
    clean_data.loc[i, "Email"] =  prenom.lower() + "." + nom.lower() + email_template
```

```python
# La colonne "Accepted"
clean_data["Accepted"] = None
for i in range(len(clean_data)):
    if (clean_data.loc[i, "age"] <= 21) and (clean_data.loc[i, "Ville_Naissance"] == "FES" or clean_data.loc[i, "Ville_Naissance"] == "MEKNES"):
        clean_data.loc[i, "Accepted"] = "YES"
    else:
        clean_data.loc[i, "Accepted"] = "NO"
```
